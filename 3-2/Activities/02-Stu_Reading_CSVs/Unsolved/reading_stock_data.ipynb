## 2. Create a DataFrame by reading in a CSV file

```python
# set the file path
filepath = Path('../Resources/shopify_stock_data.csv')
# create a Pandas DataFrame from a csv file
file_dataframe = pd.read_csv(filepath)
print(type(file_dataframe))
```
