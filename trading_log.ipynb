```python
# -*- coding: utf-8 -*-
"""
Student Do: Trading Log.

Trading Log
In this activity, you will use lists to create a trading log that tracks profits and losses for each market day of the month. 
You will iterate over the list to calculate the highest and lowest profit and loss days. 
The goal of this activity is to use for loops and lists in order to track metrics related to trading performance.

Background
Karen, an equity trader, has been tracking her profits and losses over the course of the last 20 business days. 
She wants to quickly analyze her ongoing performance as she continues to log her profits and losses every day. 
Understanding her profits and losses each day will allow Karen to evaluate her performance for all trades monthly, weekly, and over time. 
Help Karen create a program to analyze her results.


This script demonstrates how to perform basic analysis of trading profits/losses
over the course of a month (20 business days).
"""

# @TODO: Initialize the metric variables
count = 0
total = 0
minimum = 0
maximum = 0
mean = 0



# @TODO: Initialize lists to hold profitable and unprofitable day profits/losses
profit_pnl = []
loss_pnl = []



# List of trading profits/losses
trading_pnl = [ -224,  352, 252, 354, -544,
                -650,   56, 123, -43,  254,
                 325, -123,  47, 321,  123,
                 133, -151, 613, 232, -311 ]

# @TODO: Iterate over each element of the list
for pnl in trading_pnl:
    
    # @TODO: Cumulatively sum up the total and count
    count += 1
    total += pnl

    # @TODO: Write logic to determine minimum and maximum values
    if pnl < minimum:
        minimum = pnl
    elif pnl > maximum:
        maximum = pnl
        

    # @TODO: Write logic to determine profitable vs. unprofitable days
    if pnl >= 0:
        profit_pnl.append(pnl)
    else:
        loss_pnl.append(pnl)

        
# @TODO: Calculate the average
mean = round((total / count), 2)


# @TODO: Calculate count metrics
profit_count = len(profit_pnl)
loss_count = len(loss_pnl)


# @TODO: Calculate percentage metrics
profit_percent = round((profit_count / count * 100), 2)
loss_percent = round((loss_count /count *100), 2)


# @TODO: Print out the summary statistics
print("---------------------Summary Statistics------------------------")
print(f"Total trading days: {count} days")
print(f"Net profit: ${total}")
print(f"The largest profit of all time: ${maximum}")
print(f"The largest loss of all time: ${minimum}")
print(f"The avarage of the profit & loss: ${mean}")
print(f"Profitable days: {profit_count} days, accounts for {profit_percent}%")
print(f"unprofitable days: {loss_count} days, accounts for {loss_percent}%")

```
