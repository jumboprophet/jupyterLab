```python
# declair final_result as a dict
final_result = {}

# Find out the ending_weekly_close and beginning_weekly_close
for ticker,ticker_data in historical_stock_data.items():
    ending_weekly_close = ticker_data[-1][-1]
    begining_weekly_close = ticker_data[0][-1]
    
    # Calculate the ticker_weekly_close in %
    # ticker_weekly_close = ending_week_close - beginning_week_close) / beginning_week_close * 100. 
    # Round to the nearest two decimal places.
    ticker_weekly_close = round((ending_weekly_close - begining_weekly_close) / begining_weekly_close * 100 , 2)
    
    # add result to final_result
    final_result[ticker] = ticker_weekly_close

# print out the final_result
print (final_result)
```
