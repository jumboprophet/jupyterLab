```python
# -*- coding: utf-8 -*-
"""
Student Activity: Market Capitalization.

This script showcases the use of Python Dicts to determine the
bank names associated with the corresponding market cap ranges.

Market Capitalization
In this activity, you will create a dictionary, and then update, remove, 
and extract values in/from the dictionary.

Background
Sam wants to categorize banks by their market capitalizations, which is the total dollar market value of a company's outstanding shares. 
Because she wants to know the relationship between a certain bank and its market capitalization, Sam uses a dictionary to index bank names to the value of its market cap.

Sam needs to make a few changes to her dictionary of bank market caps because she noticed a few errors and omissions. 
Help Sam fix her dictionary of bank market caps, and even better, help her group the banks by their corresponding market capitalization tier.

# Change the market cap for Royal Bank of Canada to 170.
# Add a new bank Scotiabank to the dictionary and set the market cap to 33.
# Delete Montreal Trust Company from the dictionary, as it is a depreciated bank acquired by Scotiabank in 1994.

"""

# Banks and Market Caps
#-----------------------
# National Bank of Canada: 327
# Toronto-Dominion Bank: 302
# Royal Bank of Canada: 173
# Wells Fargo: 273
# Goldman Sachs: 87
# Morgan Stanley: 72
# Canadian Imperial Bank of Commerce: 83
# TD Bank: 108
# Bank of Montreal: 67
# Capital One: 47
# FNB Corporation: 4
# Laurentian Bank of Canada: 3
# Ally Financial: 12
# Montreal Trust Company: 145
# Canadian Western Bank: .97

# @TODO: Initialize a dictionary of banks and market caps (in billions)

bank_market_caps = {
        "National Bank of Canda" : 327,
        "Toronto-Dominion Bank" : 302,
        "Royal Bank of Canada" : 173,
        "Wells Fargo" : 273,
        "Goldman Sachs" : 87,
        "Morgan Stanley" : 72,
        "Canadian Imperial Bank of Commerce" : 83,
        "TD Bank" : 108,
        "Bank of Montreal" : 67,
        "Capital One" : 47,
        "FNB Corporation" : 4,
        "Laurentian Bank of Canada" : 3,
        "Ally Financial" : 12,
        "Montreal Trust Company" : 145,
        "Canadian Western Bank" : .97,
}


# @TODO: Change the market cap for 'Royal Bank of Canada'
bank_market_caps["Royal Bank of Canada"] = 170


# @TODO: Add a new bank and market cap pair
bank_market_caps["Scotiabank"] = 33


# @TODO: Remove a bank from the dictionary
del bank_market_caps["Montreal Trust Company"]




```

```python
# Total market capitalization
# Total number of banks
# Average market capitalization
# Largest bank in market capitalization
# Smallest bank in market capitalization

total_cap = 0
count_bank = 0
ave_cap = 0
min_key = ""
max_key = ""
min_bank = 0
max_bank = 0

# Initializing the lists

# we can use nested list and dictionary(list of dictionary) to make mega_cap etc. a list, each element of that list are a pair of key and value.
mega_caps = {}
large_caps = {}
mid_caps = {}
small_caps = {}

for k in bank_market_caps:
    total_cap += bank_market_caps[k]
    count_bank += 1
    
    # Calculate the min market capitalization
    if min_bank == 0:
        min_bank = bank_market_caps[k]
        min_key = k
    elif min_bank > bank_market_caps[k]:
        min_bank = bank_market_caps[k]
        min_key = k
    
    # Calculate the max market capitalization
    if max_bank == 0:
        max_bank = bank_market_caps[k]
        max_key = k
    elif max_bank < bank_market_caps[k]:
        max_bank = bank_market_caps[k]
        max_key = k
    
    # Group those banks
    if bank_market_caps[k] >= 300:
        mega_caps[k] = bank_market_caps[k]
    elif bank_market_caps[k] >= 10 and bank_market_caps[k] < 300:
        large_caps[k] = bank_market_caps[k]
    elif bank_market_caps[k] >= 2 and bank_market_caps[k] < 10:
        mid_caps[k] = bank_market_caps[k]
    else:
        small_caps[k] = bank_market_caps[k]

ave_cap = total_cap / count_bank

print("The total capitalization of all banks is: ", total_cap)
print("The total number of banks is: ", count_bank)
print("The average capitalization of all banks is: ", round(ave_cap,2))
print(f"{min_key} has the minimum market capitalization at {min_bank}")
print(f"{max_key} has the maximum market capitalization at {max_bank}")
print("\n")
print(f"mega cap firms and corresponding values are: {mega_caps}")
print(f"larege cap firms and corresponding values are: {large_caps}")
print(f"mid cap firms and corresponding values are: {mid_caps}")
print(f"small cap firms and corresponding values are: {small_caps}")
```
