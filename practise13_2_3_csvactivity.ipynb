```python
# -*- coding: utf-8 -*-
"""Student Do: Sales Analysis.

To calculate customer revenue averages as per the data in 'sales.csv'

This script will use the Pathlib library to set the file path,
use the csv library to read in the file, and iterate over each
row of the file to calculate customer sales averages.
"""

# @TODO: Import the pathlib and csv library
from pathlib import Path
import csv


# @TODO: Set the file path
csv_path = Path('./Resource/sales.csv')

# Initialize list of records
records = []

# Initialize a dictionary called analysis - key: name && value: count and revenue
analysis = {}

# @TODO: Open the csv file as an object
with open(csv_path, 'r') as csv_file:

    # @TODO:
    # Pass in the csv file to the csv.reader() function
    # (with ',' as the delmiter/separator) and return the csvreader object
    csvreader = csv.reader(csv_file, delimiter = ',')

    # @TODO: Read the header row，把第一行header剥离csvreader
    header = next(csvreader)

    # @TODO: Print the header,CSVreader读出来的每一行都是一个list
    print(type(header))
    print(header, "<--------------- This is the header row")

    # @TODO: Append the column 'Average' to the header
    header.append('Average')
    
    # @TODO: Append the header to the list of records
    records.append(header)

    # @TODO: Read each row of data after the header, 循环除了header以外的csvreader，即数据
    for row in csvreader:
        # @TODO: Print the row
        print(row)
        # @TODO:
        # Set the 'name', 'count', 'revenue' variables for better
        # readability, convert strings to ints for numerical calculations
        name = row[0]
        count = int(row[1])
        revenue = float(row[2])


        # @TODO: Calculate the average (round to the nearest 2 decimal places)
        average = round((revenue/count),2)

        # @TODO: Append the average to the row
        row.append(average)
        # @TODO: Append the row to the list of records
        records.append(row)
        
                
        # Creating the analysis{} by looking up the name
        # if not already in the analysis{}, then add new record to the analysis{}
        # keys() is to display the keys of a dictionary
        if name not in analysis.keys():
            analysis[name] = {"count" : count, "revenue" : revenue}
        else:
            analysis[name]["count"] += count
            analysis[name]["revenue"] += revenue

# @TODO: Set the path for the output.csv
output_path = Path('./Resource/sales_new.csv')
# @TODO:
# Open the output path as a file and pass into the 'csv.writer()' function
# Set the delimiter/separater as a ','
with open(output_path, 'w') as csv_file:
    csvwriter = csv.writer(csv_file, delimiter = ',')
    # @TODO:
    # Loop through the list of records and write every record to the
    # output csv file
    for line in records:
        csvwriter.writerow(line)
        
# @TODO: Set the path for the output.csv
agg_path = Path('./Resource/sales_agg.csv')

# Create header for the new list for file writer
new_header = ["Name", "Count", "Revenue", "Average"]
# @TODO:
# Open the output path as a file and pass into the 'csv.writer()' function
# Set the delimiter/separater as a ','
with open(agg_path, 'w') as csv_file:
    csvwriter = csv.writer(csv_file, delimiter = ',')
    
    csvwriter.writerow(new_header)
    # @TODO:
    # Loop through the analysis{} and write every record to the
    # output csv file
    for name in analysis:
        csvwriter.writerow(
            [
                name,
                analysis[name]["count"],
                analysis[name]["revenue"],
                round((analysis[name]["revenue"]/analysis[name]["count"]),2)
            ]
        )
```
