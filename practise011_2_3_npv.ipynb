```python
# -*- coding: utf-8 -*-
"""Student Activity: Financial Analysis using NPV.

In this activity, you will conduct financial analysis on three project scenarios: 
conservative, neutral, and aggressive. 
Each scenario requires an initial investment to generate the resulting cash flows 
for the following four years.


This script will choose the optimal project scenario to
undertake based on max NPV values.
"""

# @TODO: Import the NumPy library
import numpy_financial as npf

# Discount Rate
discount_rate = .1

# Initial Investment, Cash Flow 1, Cash Flow 2, Cash Flow 3, Cash Flow 4
cash_flows_conservative = [-1000, 400, 400, 400, 400]
cash_flows_neutral = [-1500, 600, 600, 600, 600]
cash_flows_aggressive = [-2250, 800, 800, 800, 800]

# @TODO: Initialize dictionary to hold NPV return values
npv_dict = {}

# @TODO: Calculate the NPV for each scenario
npv_conservative = npf.npv(discount_rate, cash_flows_conservative)
npv_neutral = npf.npv(discount_rate, cash_flows_neutral)
npv_aggressive = npf.npv(discount_rate, cash_flows_aggressive)

# @TODO: Manually Choose the project with the highest NPV value
print(npv_conservative)
print(npv_neutral)
print(npv_aggressive)
```

```python
# To put npvs into the dictionary
npv_dict["npv_conservative"] = npv_conservative
npv_dict["npv_neutral"] = npv_neutral
npv_dict["npv_aggressive"] = npv_aggressive
```

```python
npv_dict
```

```python
# Initialize the max key and value
max_value = 0.00
max_key = ""

# print out the dict and find out the max key and value
for key, value in npv_dict.items():
    print(f"{key}: {value}")
    
    if (max_value == 0):
        max_value = value
        max_key = key
    elif (max_value < value):
        max_value = value
        max_key = key
    
print()
print(f"The max npv is generated by the project of {max_key} with the value at {max_value}.")
```
